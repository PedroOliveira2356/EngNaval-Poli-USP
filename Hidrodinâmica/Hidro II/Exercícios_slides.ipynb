Exercício (1): Desenhe as linhas de corrente de uma fonte/sorvedouro localizada no ponto (0,0) para diferentes valores de $m$

$$\psi_S(x, y) = \pm \frac{m}{2\pi} \text{atan}\left(\frac{y - y_0}{x - x_0}\right) + cte$$

```python
import numpy as np
import matplotlib.pyplot as plt

# Função da linha de corrente
def psi_s(x, y, m, x0=0, y0=0):
    """
    Função de corrente para fonte (m > 0) ou sorvedouro (m < 0)
    psi_s = (m/2π) * atan((y-y0)/(x-x0))
    """
    return (m / (2 * np.pi)) * np.arctan2(y - y0, x - x0)

# Configuração da grade
x = np.linspace(-5, 5, 400)
y = np.linspace(-5, 5, 400)
X, Y = np.meshgrid(x, y)

# Diferentes valores de m para testar
valores_m = [10, 5, -5, -10]
cores = ['red', 'orange', 'blue', 'darkblue']
labels = ['Fonte forte (m=10)', 'Fonte fraca (m=5)', 
          'Sorvedouro fraco (m=-5)', 'Sorvedouro forte (m=-10)']

# Criar figura com subplots
fig, axes = plt.subplots(2, 2, figsize=(14, 12))
axes = axes.flatten()

for idx, (m, cor, label) in enumerate(zip(valores_m, cores, labels)):
    ax = axes[idx]
    
    # Calcular função de corrente
    PSI = psi_s(X, Y, m)
    
    # Plotar linhas de corrente
    if m > 0:
        # Fonte: linhas radiais saindo
        niveis = np.linspace(-np.pi*m/(2*np.pi), np.pi*m/(2*np.pi), 24)
    else:
        # Sorvedouro: linhas radiais entrando
        niveis = np.linspace(-np.pi*abs(m)/(2*np.pi), np.pi*abs(m)/(2*np.pi), 24)
    
    contour = ax.contour(X, Y, PSI, levels=niveis, colors=cor, linewidths=1.5)
    
    # Adicionar setas para indicar direção do fluxo
    if m > 0:
        # Fonte: setas para fora
        ax.quiver(0, 0, 1, 0, color='black', scale=5, width=0.01, 
                 headwidth=3, headlength=4, zorder=5)
        ax.text(1.5, 0.3, 'Saída', fontsize=10, fontweight='bold')
    else:
        # Sorvedouro: setas para dentro
        ax.quiver(2, 0, -1, 0, color='black', scale=5, width=0.01, 
                 headwidth=3, headlength=4, zorder=5)
        ax.text(1.5, 0.3, 'Entrada', fontsize=10, fontweight='bold')
    
    # Marcar o ponto singular (fonte/sorvedouro)
    ax.plot(0, 0, 'ko', markersize=10, label='Singularidade (x₀, y₀)', zorder=10)
    
    ax.set_xlabel('x', fontsize=12)
    ax.set_ylabel('y', fontsize=12)
    ax.set_title(label, fontsize=13, fontweight='bold')
    ax.grid(True, alpha=0.3)
    ax.set_aspect('equal')
    ax.legend(loc='upper right')
    ax.set_xlim(-5, 5)
    ax.set_ylim(-5, 5)

plt.suptitle('Linhas de Corrente: ψ(x,y) = (m/2π)·atan((y-y₀)/(x-x₀))', 
             fontsize=16, fontweight='bold', y=0.995)
plt.tight_layout()
plt.show()

# Criar uma figura adicional comparando todos os casos
fig2, ax2 = plt.subplots(1, 1, figsize=(10, 10))

for m, cor, label in zip(valores_m, cores, labels):
    PSI = psi_s(X, Y, m)
    niveis = np.linspace(-np.pi*abs(m)/(2*np.pi), np.pi*abs(m)/(2*np.pi), 20)
    ax2.contour(X, Y, PSI, levels=niveis, colors=cor, linewidths=1.2, 
                alpha=0.7, label=label)

ax2.plot(0, 0, 'ko', markersize=12, label='Singularidade', zorder=10)
ax2.set_xlabel('x', fontsize=14)
ax2.set_ylabel('y', fontsize=14)
ax2.set_title('Comparação de Linhas de Corrente para Diferentes Valores de m', 
              fontsize=14, fontweight='bold')
ax2.grid(True, alpha=0.3)
ax2.set_aspect('equal')
ax2.legend(loc='upper right', fontsize=11)
ax2.set_xlim(-5, 5)
ax2.set_ylim(-5, 5)

plt.tight_layout()
plt.show()

print("Interpretação física:")
print("=" * 60)
print("• m > 0: FONTE - fluido sai radialmente do ponto (x₀, y₀)")
print("• m < 0: SORVEDOURO - fluido entra radialmente no ponto (x₀, y₀)")
print("• |m| maior: escoamento mais intenso")
print("• Linhas de corrente: radiais a partir da singularidade")
print("• As linhas são igualmente espaçadas angularmente")
```

Exercício (3): Desenhe as linhas de corrente de vórtice localizado no ponto (0,0) para diferentes valores de $\gamma$

```python
import numpy as np
import matplotlib.pyplot as plt

# Função da linha de corrente do vórtice
def psi_v(x, y, gamma, x0=0, y0=0):
    """
    Função de corrente para vórtice
    psi_v = (gamma/2π) * ln(sqrt((x-x0)² + (y-y0)²))
    """
    r = np.sqrt((x - x0)**2 + (y - y0)**2)
    # Evitar log(0) no ponto singular
    r = np.where(r < 0.01, 0.01, r)
    return (gamma / (2 * np.pi)) * np.log(r)

# Configuração da grade
x = np.linspace(-5, 5, 400)
y = np.linspace(-5, 5, 400)
X, Y = np.meshgrid(x, y)

# Diferentes valores de gamma para testar
valores_gamma = [10, 5, -5, -10]
cores = ['red', 'orange', 'blue', 'darkblue']
labels = ['Vórtice anti-horário forte (γ=10)', 
          'Vórtice anti-horário fraco (γ=5)', 
          'Vórtice horário fraco (γ=-5)', 
          'Vórtice horário forte (γ=-10)']

# Criar figura com subplots
fig, axes = plt.subplots(2, 2, figsize=(14, 12))
axes = axes.flatten()

for idx, (gamma, cor, label) in enumerate(zip(valores_gamma, cores, labels)):
    ax = axes[idx]
    
    # Calcular função de corrente
    PSI = psi_v(X, Y, gamma)
    
    # Plotar linhas de corrente (círculos concêntricos)
    niveis = np.linspace(PSI.min() + 0.5, PSI.max() - 0.5, 25)
    contour = ax.contour(X, Y, PSI, levels=niveis, colors=cor, linewidths=1.5)
    
    # Adicionar setas para indicar direção do fluxo
    theta_arrows = np.array([0, np.pi/2, np.pi, 3*np.pi/2])
    r_arrow = 2.5
    
    for theta in theta_arrows:
        x_arr = r_arrow * np.cos(theta)
        y_arr = r_arrow * np.sin(theta)
        
        if gamma > 0:
            # Anti-horário
            dx = -np.sin(theta) * 0.8
            dy = np.cos(theta) * 0.8
        else:
            # Horário
            dx = np.sin(theta) * 0.8
            dy = -np.cos(theta) * 0.8
        
        ax.arrow(x_arr, y_arr, dx, dy, head_width=0.3, head_length=0.2, 
                fc='black', ec='black', linewidth=2, zorder=5)
    
    # Marcar o ponto singular (vórtice)
    ax.plot(0, 0, 'ko', markersize=10, label='Vórtice (x₀, y₀)', zorder=10)
    
    # Adicionar texto indicando rotação
    if gamma > 0:
        ax.text(0, -4.2, '↺ Anti-horário', fontsize=12, fontweight='bold', 
                ha='center', color=cor)
    else:
        ax.text(0, -4.2, '↻ Horário', fontsize=12, fontweight='bold', 
                ha='center', color=cor)
    
    ax.set_xlabel('x', fontsize=12)
    ax.set_ylabel('y', fontsize=12)
    ax.set_title(label, fontsize=13, fontweight='bold')
    ax.grid(True, alpha=0.3)
    ax.set_aspect('equal')
    ax.legend(loc='upper right')
    ax.set_xlim(-5, 5)
    ax.set_ylim(-5, 5)

plt.suptitle('Linhas de Corrente: ψ(x,y) = (γ/2π)·ln(√[(x-x₀)² + (y-y₀)²])', 
             fontsize=16, fontweight='bold', y=0.995)
plt.tight_layout()
plt.show()

# Criar uma figura adicional comparando todos os casos
fig2, ax2 = plt.subplots(1, 1, figsize=(10, 10))

for gamma, cor, label in zip(valores_gamma, cores, labels):
    PSI = psi_v(X, Y, gamma)
    niveis = np.linspace(PSI.min() + 0.5, PSI.max() - 0.5, 20)
    ax2.contour(X, Y, PSI, levels=niveis, colors=cor, linewidths=1.2, 
                alpha=0.7, label=label)

# Adicionar setas de direção
theta_arrows = np.linspace(0, 2*np.pi, 8, endpoint=False)
r_arrow = 3.0

for theta in theta_arrows:
    x_arr = r_arrow * np.cos(theta)
    y_arr = r_arrow * np.sin(theta)
    # Seta anti-horária (para gamma positivo)
    dx = -np.sin(theta) * 0.5
    dy = np.cos(theta) * 0.5
    ax2.arrow(x_arr, y_arr, dx, dy, head_width=0.2, head_length=0.15, 
            fc='darkred', ec='darkred', linewidth=1.5, alpha=0.7, zorder=5)

ax2.plot(0, 0, 'ko', markersize=12, label='Vórtice', zorder=10)
ax2.text(0, -4.5, '↺ γ>0: Anti-horário  |  ↻ γ<0: Horário', 
         fontsize=12, fontweight='bold', ha='center', 
         bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.5))

ax2.set_xlabel('x', fontsize=14)
ax2.set_ylabel('y', fontsize=14)
ax2.set_title('Comparação de Linhas de Corrente para Diferentes Valores de γ', 
              fontsize=14, fontweight='bold')
ax2.grid(True, alpha=0.3)
ax2.set_aspect('equal')
ax2.legend(loc='upper right', fontsize=10)
ax2.set_xlim(-5, 5)
ax2.set_ylim(-5, 5)

plt.tight_layout()
plt.show()

print("Interpretação física:")
print("=" * 70)
print("• γ > 0: VÓRTICE ANTI-HORÁRIO - fluido circula no sentido anti-horário")
print("• γ < 0: VÓRTICE HORÁRIO - fluido circula no sentido horário")
print("• |γ| maior: rotação mais intensa")
print("• Linhas de corrente: círculos concêntricos centrados em (x₀, y₀)")
print("• Velocidade tangencial: v_θ = γ/(2πr) - decresce com 1/r")
print("• Velocidade radial: v_r = 0 - não há fluxo radial")
print("=" * 70)
```

Exercício (5): Desenhe as linhas de corrente associadas aos seguintes potenciais de velocidades para diferentes valores de $U$, $a$, $m$:

(a) $$\phi(x, y) = Ux + \frac{m}{2\pi} \ln\left(\sqrt{(x-a)^2 + y^2}\right)$$

(b) $$\phi(x, y) = Ux + \frac{m}{2\pi} \ln\left(\sqrt{(x+a)^2 + y^2}\right) - \frac{m}{2\pi} \ln\left(\sqrt{(x-a)^2 + y^2}\right)$$

```python
import numpy as np
import matplotlib.pyplot as plt

# ============================================================================
# FUNÇÕES DOS POTENCIAIS DE VELOCIDADE
# ============================================================================

def phi_a(x, y, U, m, a):
    """
    Item (a): Escoamento uniforme + fonte
    φ(x,y) = Ux + (m/2π)ln(√[(x-a)² + y²])
    """
    r = np.sqrt((x - a)**2 + y**2)
    r = np.where(r < 0.01, 0.01, r)  # Evitar singularidade
    return U * x + (m / (2 * np.pi)) * np.log(r)

def phi_b(x, y, U, m, a):
    """
    Item (b): Escoamento uniforme + fonte em (-a,0) + sorvedouro em (a,0)
    φ(x,y) = Ux + (m/2π)ln(√[(x+a)² + y²]) - (m/2π)ln(√[(x-a)² + y²])
    """
    r1 = np.sqrt((x + a)**2 + y**2)
    r2 = np.sqrt((x - a)**2 + y**2)
    r1 = np.where(r1 < 0.01, 0.01, r1)
    r2 = np.where(r2 < 0.01, 0.01, r2)
    return U * x + (m / (2 * np.pi)) * np.log(r1) - (m / (2 * np.pi)) * np.log(r2)

def psi_a(x, y, U, m, a):
    """Função de corrente para item (a)"""
    return U * y + (m / (2 * np.pi)) * np.arctan2(y, x - a)

def psi_b(x, y, U, m, a):
    """Função de corrente para item (b)"""
    return U * y + (m / (2 * np.pi)) * np.arctan2(y, x + a) - (m / (2 * np.pi)) * np.arctan2(y, x - a)

# ============================================================================
# PARÂMETROS PARA SIMULAÇÃO
# ============================================================================

# Configuração da grade
x = np.linspace(-5, 5, 500)
y = np.linspace(-4, 4, 400)
X, Y = np.meshgrid(x, y)

# Diferentes conjuntos de parâmetros
parametros = [
    {'U': 1, 'm': 2, 'a': 1, 'label': 'U=1, m=2, a=1'},
    {'U': 2, 'm': 3, 'a': 1, 'label': 'U=2, m=3, a=1'},
    {'U': 1, 'm': 4, 'a': 1.5, 'label': 'U=1, m=4, a=1.5'},
]

# ============================================================================
# ITEM (A): ESCOAMENTO UNIFORME + FONTE
# ============================================================================

fig1, axes1 = plt.subplots(len(parametros), 1, figsize=(14, 5*len(parametros)))
if len(parametros) == 1:
    axes1 = [axes1]

for idx, params in enumerate(parametros):
    U, m, a = params['U'], params['m'], params['a']
    ax = axes1[idx]
    
    # Calcular função de corrente
    PSI_A = psi_a(X, Y, U, m, a)
    
    # Plotar linhas de corrente
    niveis = np.linspace(PSI_A.min() * 0.8, PSI_A.max() * 0.8, 30)
    contour = ax.contour(X, Y, PSI_A, levels=niveis, colors='blue', linewidths=1.2)
    ax.clabel(contour, inline=True, fontsize=8, fmt='%1.1f')
    
    # Marcar a fonte
    ax.plot(a, 0, 'ro', markersize=12, label=f'Fonte em ({a}, 0)', zorder=10)
    
    # Adicionar setas do escoamento uniforme
    ax.arrow(-4, 3, 1, 0, head_width=0.2, head_length=0.2, 
             fc='black', ec='black', linewidth=2)
    ax.text(-3, 3.5, f'U = {U}', fontsize=12, fontweight='bold')
    
    # Configurações do gráfico
    ax.set_xlabel('x', fontsize=12)
    ax.set_ylabel('y', fontsize=12)
    ax.set_title(f'Item (a): Escoamento Uniforme + Fonte | {params["label"]}', 
                fontsize=13, fontweight='bold')
    ax.grid(True, alpha=0.3)
    ax.set_aspect('equal')
    ax.legend(loc='upper right')
    ax.set_xlim(-5, 5)
    ax.set_ylim(-4, 4)

plt.suptitle('φ(x,y) = Ux + (m/2π)ln(√[(x-a)² + y²])', 
             fontsize=16, fontweight='bold', y=0.995)
plt.tight_layout()
plt.show()

# ============================================================================
# ITEM (B): ESCOAMENTO UNIFORME + FONTE + SORVEDOURO (DIPOLO)
# ============================================================================

fig2, axes2 = plt.subplots(len(parametros), 1, figsize=(14, 5*len(parametros)))
if len(parametros) == 1:
    axes2 = [axes2]

for idx, params in enumerate(parametros):
    U, m, a = params['U'], params['m'], params['a']
    ax = axes2[idx]
    
    # Calcular função de corrente
    PSI_B = psi_b(X, Y, U, m, a)
    
    # Plotar linhas de corrente
    niveis = np.linspace(PSI_B.min() * 0.8, PSI_B.max() * 0.8, 30)
    contour = ax.contour(X, Y, PSI_B, levels=niveis, colors='darkgreen', linewidths=1.2)
    ax.clabel(contour, inline=True, fontsize=8, fmt='%1.1f')
    
    # Marcar fonte e sorvedouro
    ax.plot(-a, 0, 'ro', markersize=12, label=f'Fonte em ({-a}, 0)', zorder=10)
    ax.plot(a, 0, 'bo', markersize=12, label=f'Sorvedouro em ({a}, 0)', zorder=10)
    
    # Adicionar setas do escoamento uniforme
    ax.arrow(-4, 3, 1, 0, head_width=0.2, head_length=0.2, 
             fc='black', ec='black', linewidth=2)
    ax.text(-3, 3.5, f'U = {U}', fontsize=12, fontweight='bold')
    
    # Configurações do gráfico
    ax.set_xlabel('x', fontsize=12)
    ax.set_ylabel('y', fontsize=12)
    ax.set_title(f'Item (b): Escoamento Uniforme + Dipolo | {params["label"]}', 
                fontsize=13, fontweight='bold')
    ax.grid(True, alpha=0.3)
    ax.set_aspect('equal')
    ax.legend(loc='upper right')
    ax.set_xlim(-5, 5)
    ax.set_ylim(-4, 4)

plt.suptitle('φ(x,y) = Ux + (m/2π)ln(√[(x+a)² + y²]) - (m/2π)ln(√[(x-a)² + y²])', 
             fontsize=14, fontweight='bold', y=0.995)
plt.tight_layout()
plt.show()

# ============================================================================
# INTERPRETAÇÃO FÍSICA
# ============================================================================

print("\n" + "=" * 80)
print("INTERPRETAÇÃO FÍSICA:")
print("=" * 80)
print("\nItem (a): ESCOAMENTO UNIFORME + FONTE")
print("-" * 80)
print("• Superposição de escoamento uniforme U (da esquerda para direita)")
print("• Fonte de intensidade m localizada em (a, 0)")
print("• As linhas de corrente se afastam da fonte")
print("• Representa escoamento ao redor de um corpo (meio-corpo de Rankine)")
print()

print("Item (b): ESCOAMENTO UNIFORME + DIPOLO")
print("-" * 80)
print("• Superposição de escoamento uniforme U")
print("• Fonte em (-a, 0) e sorvedouro em (a, 0)")
print("• Forma um dipolo: fonte + sorvedouro")
print("• As linhas se fecham formando padrão oval")
print("• Representa escoamento ao redor de um cilindro circular")
print("• Quando a→0 com m·a=const, obtém-se o dipolo puro")
print("=" * 80)
```

Para um cilíndro de raio $R$ imerso num fluido de escoamento uniforme $U_\infty$, temos:

$$\psi(r,\theta) = U_\infty \sin \theta \left( r - \dfrac{R^2}{r} \right)$$

```python
import numpy as np
import matplotlib.pyplot as plt

# Parâmetros
U = 1.0     # Velocidade do escoamento livre
R = 1.0     # Raio do cilindro

# Malha cartesiana
x = np.linspace(-3*R, 3*R, 200)
y = np.linspace(-3*R, 3*R, 200)
X, Y = np.meshgrid(x, y)

# Conversão para coordenadas polares
r = np.sqrt(X**2 + Y**2)
theta = np.arctan2(Y, X)

# Evitar divisão por zero no centro
r[r < R] = np.nan

# Componentes de velocidade
u_r = U * (1 - (R**2)/(r**2)) * np.cos(theta)
u_theta = -U * (1 + (R**2)/(r**2)) * np.sin(theta)

# Conversão para coordenadas cartesianas
u_x = u_r * np.cos(theta) - u_theta * np.sin(theta)
u_y = u_r * np.sin(theta) + u_theta * np.cos(theta)

# Plot
fig, ax = plt.subplots(figsize=(7, 7))
ax.streamplot(X, Y, u_x, u_y, color='b', density=2, linewidth=1)

# Desenha o cilindro
circle = plt.Circle((0, 0), R, color='gray', zorder=10)
ax.add_patch(circle)

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Linhas de corrente - Escoamento uniforme ao redor de um cilindro')
ax.set_aspect('equal', 'box')
plt.show()
```

Exercício: Considere o painel de fontes de intensidade constante $\sigma=2m/s$

(1) Considere o ponto $(2,1)$ como vértice inicial do painel e calcule as componentes de velocidade induzidas no ponto $P(1,3)$

(2) Refaça os cálculos considerando agora o ponto $(4,2)$ como vértice inercial e verifique que as velocidades em $P$ não se alteram

```python
import math

# intensidade da fonte [m/s]
sigma = 2

# Coordenadas do painel
xa, ya = 2, 1   # ponto inicial do painel
xb, yb = 4, 2   # ponto final do painel

# Coordenadas ponto P
xp = 1; yp = 3

# Comprimento e ângulo do painel
a = 0
b = math.sqrt((xb - xa)**2 + (yb - ya)**2)
beta = math.atan2(yb - ya, xb - xa)


# Coordenadas locais do ponto P em relação ao painel
X = (xp - xa)*math.cos(beta) + (yp - ya)*math.sin(beta)
Y = -(xp - xa)*math.sin(beta) + (yp - ya)*math.cos(beta)

# Componentes da velocidade induzida pela fonte no ponto P no refêrencial do painel
U = (sigma / (4*math.pi)) * math.log((X**2 + Y**2) / ((X-b)**2 + Y**2))
V = (sigma / (2*math.pi)) * (math.atan2(X,Y) - math.atan2(X-b,Y))

# Componentes da velocidade no referencial global
u = U * math.cos(beta) - V * math.sin(beta)
v = U * math.sin(beta) + V * math.cos(beta)

print(f"Componente u da velocidade no ponto P: {u:.4f} m/s")
print(f"Componente v da velocidade no ponto P: {v:.4f} m/s")
```

```python
# Características ambientais
rho = 1025      # Densidade da água do mar [kg/m³]
g = 9.81        # Aceleração da gravidade [m/s²]

# Características da FPSO
Nabla = 214160      # Deslocamento [m³]
Aw = 15300          # Área de linha d'água [m²]
kxx = 17.82         # Raio de giração transversal [m]
GM_T = 8.34         # Metacentro transversal [m]


# Cálculo do período natural de heave (T_n3)
a_33 = Nabla * rho                  # Massa adicional (33) [kg]
T_n3 = 2*math.pi*math.sqrt((rho*Nabla + a_33)/(rho*g*Aw))

# Cálculo do período natural de roll (T_n4)
I_xx = rho * Nabla * kxx**2         # Momento de inércia do corpo em roll [kg·m²]
a_44 = 0.5*I_xx
T_n4 = (2*math.pi)/math.sqrt((rho * Nabla * g * GM_T)/(I_xx + a_44))

print(f"Período natural de heave (T_n3): {T_n3:.2f} s")
print(f"Período natural de roll (T_n4): {T_n4:.2f} s")
```

## Slide 17 - A Fórmula de Morison

Exercício: estime a amplitude da força horizontal de arrasto viscoso que age sobre uma coluna vertical fixa de seção circular em águas rasas, considerando os dados abaixo.

```python
import numpy as np
from scipy.optimize import fsolve
from scipy.integrate import quad

# Características da coluna vertical
D = 0.3         # Diâmetro [m]
h = 5.0         # Altura [m]
A = 0.25        # Área da seção transversal [m²]
T = 15.0        # Período da onda [s]

# Carcaterísticas ambientais
rho = 1025      # Densidade da água do mar [kg/m³]
g = 9.81        # Aceleração da gravidade [m/s²]
nu = 1.2e-6     # Viscosidade cinemática da água do mar [m²/s]


# Cálculo de constantes úteis
omega = 2*np.pi/T

def func(k_f):
    return (omega**2) / (g*k_f*h) - k_f
k = fsolve(func, 1.0)[0]

lambda_ = 2*np.pi/k
ka = np.pi*D/lambda_
KC = ((A*T)/D) * np.sqrt(g/h)
beta = (1/nu)*((D**2)/T)

# Valores do gráfico
C_M = 1.7; C_D = 0.8


# Cálculo das componentes de velocidade e aceleração
def u_0(z, t):
    return omega*A*(np.cosh(k*(z + h)) / np.sinh(k*h)) * np.cos(omega*t)
def w_0(z, t):
    return omega*A*(np.sinh(k*(z + h)) / np.sinh(k*h)) * np.sin(omega*t)
def u_0_dot(z, t):
    return -omega**2*A*(np.cosh(k*(z + h)) / np.sinh(k*h)) * np.sin(omega*t)
def w_0_dot(z, t):
    return  omega**2*A*(np.sinh(k*(z + h)) / np.sinh(k*h)) * np.cos(omega*t)

# Força local (densidade de força) pelo modelo de Morison
def f_x(z, t):
    u = u_0(z, t)
    u_dot = u_0_dot(z, t)
    return rho*A*C_M*u_dot + 0.5*(rho*D*C_D*u*abs(u))

# Integração numérica ao longo da coluna (de z = -5 até 0)
t = 0.0
F_x, _ = quad(f_x, -5, 0, args=(t,))

print(f'F_x = {F_x:.2f} N')
```

## Slide 21 - Introdução ao tratamento estatístico do estudo de comportamento no mar

Exercício: A série temporal do registro de ondas apresentado no início desta aula é fornecida em arquivo ascii no e-disciplinas. (Atenção: primeiro são apresentados os valores do tempo de 0 a 1800 s, seguidos dos valores de elevação de ondas)

A partir deste registro, estime:

a) A amplitude média de ondas

b) A altura significativa de ondas

c) A probabilidade de a amplitude de onda exceder o valor de 1,0m

```python
import numpy as np
import matplotlib.pyplot as plt
matplotlib.style.use('ggplot')
from scipy.integrate import quad
```

```python

filename = 'ondas.txt'
data = np.loadtxt(filename, delimiter=',', skiprows=0, dtype=float)
tempo = data[:9000]
amp = data[9000:]

plt.figure(figsize=(8,4))
plt.plot(tempo, amp, 'b', linewidth=0.2)
plt.xlabel('Tempo (s)')
plt.ylabel('Elevação de onda (m)');
```

```python
# Desvio Padrao
desvpad = np.std(amp)

# Distribuicao de Rayleigh
def f(A):
    return A/(desvpad**2)*np.exp(-(A**2)/(2*desvpad**2))

x = np.linspace(0,5,1000)

plt.plot(x, f(x), 'k', linewidth=0.5)
plt.xlabel('Amplitude (m)')
plt.title('Distribuição de amplitude');
```

```python
# Amplitude media
def integrand(A):
    return A*f(A)

print(f"a) Amplitude média de ondas da distribuição de Rayleigh: {quad(integrand, 0, np.inf)[0]:.4f} m")
```

```python
a0 = 1.4823*desvpad

# Altura significativa
A_s = quad(integrand, a0, np.inf)
print('b) Altura significativa de ondas A_s =', A_s[0])
print('H_s =', A_s[0]*2)
print(f'Verificação: A_s = 2sigma => {A_s[0]} ~ {2*desvpad}')

plt.plot(x, f(x), 'k', linewidth=0.5)
plt.fill_between(x[178:], f(x)[178:], alpha = 0.8)
plt.xlabel('Amplitude (m)')
plt.title('Distribuição de amplitude');
```

```python
# Probabilidade de a amplitude de onda exceder o valor de a
a = 1
P_a = np.exp(-a**2/(2*desvpad**2))
print('c) P(A>a) =', P_a)
```

Exercício: Considere um estado de mar com altura significativa $H_S=4,0m$ e $T_P=10s$.

a) Faça um gráfico do espectro de JONSWAP desse mar;

b) Obtenha o valor de período entre zeros ascendentes;

c) Estime a probabilidade de a onda exceder uma altura de 5 metros

```python
# a) Espectro de JONSWAP
def jonswap(w):
    H_s = 4; T_p = 10; gamma = 3.3
    w_p = 2*np.pi/T_p

    if w < w_p:
        mu = 0.07
    else:
        mu = 0.09
    
    A = np.exp(-((w/w_p-1)/(mu*np.sqrt(2)))**2)

    S_zeta = (320*H_s**2)/(T_p**4*w**5) * np.exp(-1950/(T_p**4*w**4)) * gamma**A

    return S_zeta

omega = np.linspace(10**(-8),2,1000)
JS = []
for elem in omega:
    JS.append(jonswap(elem))
JS = np.array(JS)

# Espectro de JONSWAP
plt.plot(omega, JS, 'k', linewidth=0.7)
plt.xlabel('w (rad/s)')
plt.ylabel('S_zeta (m^2.s)');
```

```python
# Momento espectral de ordem k
def m_0(w):
    return w**0 * jonswap(w)

def m_2(w):
    return w**2 * jonswap(w)

m0 = quad(m_0, 0, np.inf)
m2 = quad(m_2, 0, np.inf)

# b) Periodo entre zeros ascendentes
T_z = 2*np.pi*np.sqrt(m0[0]/m2[0])
print(f'b) T_z = {T_z:.2f} s')


# c) Probabilidade da onda exceder 5 m
desvpad = np.std(JS)

a = 5
P_a = np.exp(-a**2/(2*desvpad**2))
print(f'c) P(A>a) = {100*P_a:.5f} %')
```
